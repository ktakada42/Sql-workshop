## 課題1: Popular Recommend

作成したレコメンドの説明

~~

**スコア上位100のアニメを降順に出力。**
---
列はそれぞれランキング、アニメ名、アニメ名(日本語)、スコア、スコア件数、ジャンル、放送時期、あらすじ。  
<u> *ただし、スコア件数が全体平均(約23,200件)に満たないものは除外。</u>  
    <u> *スコアが同点の場合は、スコア件数の多い方を順位が上とした。</u>

```python
%%sql
SELECT ROW_NUMBER() OVER (
    ORDER BY A.score DESC, (score_1 + score_2 + score_3 + score_4 + score_5 + score_6 + score_7 + score_8 + score_9 + score_10) DESC) AS rank,
A.name, A.japanese_name, A.score,
(score_1 + score_2 + score_3 + score_4 + score_5 + score_6 + score_7 + score_8 + score_9 + score_10) AS number_of_scores,
A.genres, A.premiered, AWS.sypnosis
FROM anime AS A INNER JOIN anime_with_synopsis AS AWS
ON A.mal_id = AWS.mal_id
WHERE (score_1 + score_2 + score_3 + score_4 + score_5 + score_6 + score_7 + score_8 + score_9 + score_10) >= 
(SELECT AVG((score_1 + score_2 + score_3 + score_4 + score_5 + score_6 + score_7 + score_8 + score_9 + score_10)) FROM Anime)
ORDER BY rank
LIMIT 100;
```

## 課題2: Item to Item Recommend

作成したレコメンドの説明

~~

**そのアニメを視聴した他のユーザーが視聴を完了したアニメを、観られた数が多い順に20個出力。**
---
列はそれぞれアニメ名、アニメ名(日本語)、平均スコア、視聴を完了したユーザー数、ジャンル、放送時期、あらすじ。  
<u> *視聴完了とは、最終話までアニメを観るか、途中で観ることをやめた状態を指す。</u>  
<u> *ユーザー数が同じ場合は、平均スコアの高い方を順位が上とした。</u>  
<u> *下記は、「天気の子」を観たユーザーへのレコメンド。</u>

```python
%%sql
SELECT ROW_NUMBER() OVER (ORDER BY watched_user DESC, average_score DESC) AS rank,
A.name, A.japanese_name, ROUND(RAI.average_score, 2) AS average_score, RAI.watched_user, A.genres, A.premiered, AWS.sypnosis
FROM anime AS A INNER JOIN (
    SELECT anime_id, COUNT(*) AS watched_user, AVG(rating) AS average_score
    FROM animelist
    WHERE user_id = ANY (SELECT user_id FROM animelist WHERE anime_id = 38826 AND watching_status IN (2, 4))
      AND anime_id <> 38826
      AND watching_status IN (2, 4)
    GROUP BY anime_id
    ORDER BY average_score DESC) AS RAI
ON A.mal_id = RAI.anime_id
INNER JOIN anime_with_synopsis AS AWS
ON A.mal_id = AWS.mal_id
ORDER BY rank
LIMIT 20;
```
